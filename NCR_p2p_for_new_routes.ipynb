```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Creating a Graph
G = nx.Graph()
```

```python
# Visualisation helper function
def view_graph(G):
    import warnings
    warnings.filterwarnings("ignore")
    # Print nodes and edges     
    print(f'Nodes in graph: {G.nodes}')
    print(f'Edges in graph: {G.edges}')
    # Create a visualisation using NetworkX and draw using matplotlib
    pos = nx.spring_layout(G)
    # Extract node attributes and display them as labels
    if len(nx.get_node_attributes(G, 'population')) != 0:
        labels = {key: key + ':' + str(value) for (key, value) in nx.get_node_attributes(G, 'population').items()}
        nx.draw(G, pos=pos, with_labels=True, labels=labels)
    else:
        nx.draw(G, pos=pos, with_labels=True)
    # Extract edge attributes and display them as labels    
    if len(nx.get_edge_attributes(G,'distance')) != 0:
        edge_labels = nx.get_edge_attributes(G, 'distance')
        nx.draw_networkx_edge_labels(G, pos=pos, edge_labels = edge_labels)
    plt.figure(figsize=(12,12)) 
    plt.show()
```

```python
#Get degree values of each terminal
degrees = dict(p2p1.degree())
nx.set_node_attributes(p2p1, name='degree', values=degrees)
```

```python
#Which terminals have the highest degree of centrality?
degree_df = pd.DataFrame(p2p1.nodes(data='degree'), columns=['node', 'degree'])
degree_df = degree_df.sort_values(by='degree', ascending=False)
degree_df
```

```python
degree_df.describe()
```

```python
#Plot degree scores of each terminal
num_nodes_to_inspect = 20
degree_df[:num_nodes_to_inspect].plot(x='node', y='degree', kind='barh').invert_yaxis()
```

```python
#Get the weighted degree of each terminal
nx.degree(p2p1, weight='Weight')
```

```python
weighted_degrees = dict(nx.degree(p2p1, weight='Weight'))
nx.set_node_attributes(p2p1, name='weighted_degree', values=weighted_degrees)
```

```python
#Which terminals have the highest weighted degree?
weighted_degree_df = pd.DataFrame(p2p1.nodes(data='weighted_degree'), columns=['node', 'weighted_degree'])
weighted_degree_df = weighted_degree_df.sort_values(by='weighted_degree', ascending=False)
weighted_degree_df
```

```python
#Plot the hbar of the weighted degree of each terminal
num_nodes_to_inspect = 20
weighted_degree_df[:num_nodes_to_inspect].plot(x='node', y='weighted_degree', color='orange', kind='barh').invert_yaxis()
```

```python
#Get the betweenness centrality of each terminal
betweenness_centrality = nx.betweenness_centrality(p2p1)
```

```python
nx.set_node_attributes(p2p1, name='betweenness', values=betweenness_centrality)
```

```python
#Which terminals have the highest betweenness?
betweenness_df = pd.DataFrame(p2p1.nodes(data='betweenness'), columns=['node', 'betweenness'])
betweenness_df = betweenness_df.sort_values(by='betweenness', ascending=False)
betweenness_df
```

```python
#Plot the hbar of the betweenness of each terminal
num_nodes_to_inspect = 20
betweenness_df[:num_nodes_to_inspect].plot(x='node', y='betweenness', color='green', kind='barh').invert_yaxis()
```

```python
#Get closeness centrality of each terminal
closeness_centrality = nx.closeness_centrality(p2p1)
```

```python
nx.set_node_attributes(p2p1, name='closeness', values=closeness_centrality)
```

```python
#Which terminals have the highest closeness value?
closeness_df = pd.DataFrame(p2p1.nodes(data='closeness'), columns=['node', 'closeness'])
closeness_df = closeness_df.sort_values(by='closeness', ascending=False)
closeness_df
```

```python
#Plot the hbar of the closeness of each terminal
num_nodes_to_inspect = 20
closeness_df[:num_nodes_to_inspect].plot(x='node', y='closeness', color='green', kind='barh').invert_yaxis()
```

```python
#Get constraint score of each terminal
constraint=nx.constraint(p2p1)
constraint
nx.set_node_attributes(p2p1, name='constraint', values=constraint)
```

```python
#Which terminals have the highest constraint?
constraint_df = pd.DataFrame(p2p1.nodes(data='constraint'), columns=['node','constraint'])
constraint_df.sort_values(by='constraint', ascending=False, inplace=False)
constraint_df
```

```python
#Plot the hbar of the cosntraint score of each terminal
num_nodes_to_inspect = 20
constraint_df[:num_nodes_to_inspect].plot(x='node', y='constraint', color='green', kind='barh').invert_yaxis()
```

```python
#List of the dataframes of centrality measures
df_list=[degree_df, weighted_degree_df,betweenness_df, closeness_df, constraint_df]
```

```python
#Merge centrality measures to one dataframe
merged_df= weighted_degree_df.merge(degree_df, how ='left', on ='node')
merged_df = merged_df.merge(betweenness_df, how ='left', on ='node')
merged_df = merged_df.merge(closeness_df, how ='left', on ='node')
merged_df = merged_df.merge(constraint_df, how ='left', on ='node')
features_df = merged_df
```
